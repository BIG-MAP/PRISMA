## Generate data to load in examples

```python
import numpy as np
import pandas as pd
import prisma.lineshapes
import bqplot.pyplot as plt
```

```python
x = np.arange(200,1500,2)
params1 = [200,650,50]
params2 = [500,730,10]
params3 = [300,900,20]
noise = 40*np.random.rand(len(x))
baseline =  3e-4*x**2

spectra_dict = {}
for n in range(20):
    params1 = [(1-0.005*n)*params1[0],params1[1],params1[2]]
    params2 = [params2[0],params2[1]+0.5*n,params2[2]]
    params3 = [(1+0.005*n)*params3[0],params3[1]+n,params3[2]]
    spectra_dict['spectrum_{}'.format(n)] = prisma.lineshapes.lorentzians(3)(x, 0, *params1, *params2, *params3) + noise + baseline
```

```python
# lor_sum = prisma.lineshapes.lorentzians(3)(x, 0, *params1, *params2, *params3)
```

```python
df = pd.DataFrame(spectra_dict, index=x)
```

```python
plt.plot(df.index, df['spectrum_0'])
plt.plot(df.index, df['spectrum_9'])
plt.plot(df.index, df['spectrum_19'])
plt.show()
```

```python
df.to_csv('synthetic_dataset.csv')
```
