## Generate data to load in examples

```python
import numpy as np
import pandas as pd
import prisma.lineshapes
import bqplot.pyplot as plt
```

```python
x = np.arange(200,1500,2)
params1 = [200,650,50]
params2 = [500,730,10]
params3 = [300,900,20]
noise = 40*np.random.rand(len(x))
baseline =  3e-4*x**2

spectra_dict = {}
for n in range(20):
    params1 = [(1-0.005*n)*params1[0],params1[1],params1[2]]
    params2 = [params2[0],params2[1]+0.5*n,params2[2]]
    params3 = [(1+0.005*n)*params3[0],params3[1]+n,params3[2]]
    spectra_dict['spectrum_{}'.format(n)] = prisma.lineshapes.lorentzians(3)(x, 0, *params1, *params2, *params3) + noise + baseline
```

```python
# lor_sum = prisma.lineshapes.lorentzians(3)(x, 0, *params1, *params2, *params3)
```

```python
df = pd.DataFrame(spectra_dict, index=x)
```

```python
df
```

```python
lor1 = prisma.lineshapes.lorentzians(1)(x, 0, *params1)
```

```python
subset_idxs = (df.index>400)&(df.index<1300)

params1 = [200,650,50]
params2 = [500,730,10]
params3 = [300,900,20]
lor1 = prisma.lineshapes.lorentzians(1)(df[subset_idxs].index, 0, *params1)
lor2 = prisma.lineshapes.lorentzians(1)(df[subset_idxs].index, 0, *params2)
lor3 = prisma.lineshapes.lorentzians(1)(df[subset_idxs].index, 0, *params3)

plt.figure(background_style={'background-color':'white'})
plt.scatter(df.index, df['spectrum_0'], colors=['gray'], default_size=10)
plt.scatter(df[subset_idxs].index, df[subset_idxs]['spectrum_0'], colors=['DodgerBlue'], default_size=10)
plt.plot(df[subset_idxs].index, lor1+3e-4*df[subset_idxs].index**2, colors=['DarkTurquoise'])
plt.plot(df[subset_idxs].index, lor2+3e-4*df[subset_idxs].index**2, colors=['DarkTurquoise'])
plt.plot(df[subset_idxs].index, lor3+3e-4*df[subset_idxs].index**2, colors=['DarkTurquoise'])
plt.plot(df[subset_idxs].index, 3e-4*df[subset_idxs].index**2, colors=['tomato'])
# plt.plot(df.iloc[(df.index>400)&(df.index<1300)].index, df['spectrum_19'])
plt.show()
```
