```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
```

```python
df = pd.read_csv(r"C:\Users\eibarc\Downloads\Processed_spectra (4).csv")
df
```

```python
fig = go.Figure()

# Plotting each array using add_trace
x_array = 'index'
col_names = df.columns.to_list()[1:]
for col in col_names:
    fig.add_trace(go.Scatter(x=df[x_array], y=df[col], mode='lines', name=col))

# Update layout (optional): Adding titles, axis labels, etc.
fig.update_layout(
    title='Tests outputs',
    xaxis=dict(title='Index'),
    yaxis=dict(title='Counts')
)

# Show the plot
fig.show()
```
